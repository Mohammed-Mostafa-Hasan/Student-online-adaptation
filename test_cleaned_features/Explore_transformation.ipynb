```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.gofplots import qqplot
from sklearn.preprocessing import LabelEncoder
```

```python
stu_data = pd.read_csv('students_adaptability_level_online_education-Copy1.csv')
```

```python
#find the feature need to transform
'''write any thing you remember here
   first using displot whose do not have normal distribution
   second find probability of very well fitted data
   thrid using pdf(probability density function)'''
```

```python
stu_data.head()
```

```python
le = LabelEncoder()

for features in stu_data.columns:
    stu_data[features]=le.fit_transform(stu_data[features].astype(str))
    
```

```python
for features in ['Class Duration','Financial Condition', 'Education Level', 'Age']:

    sns.displot(stu_data[features],kde=True)
    plt.title('shape of distplot features // {}'.format(features))
    plt.show()
```

```python
'''we see that finanicial Condition have normal distribution '
'so we need to make transformation for other feature:
'Financial Condition Education Level Age '''
for feature in ['Class Duration', 'Education Level', 'Age']:
    for i in [.5,1,2,3,4,5,6]:
        data_t = stu_data[feature]**(1/i)
        qqplot(data_t,line='s')
        plt.title("Transformation: 1/{} for// {}".format(str(i),feature))
```
